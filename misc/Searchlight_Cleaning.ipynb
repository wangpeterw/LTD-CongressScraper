```python
##untested function to concatonate csv files
##input: Path object to folder containing csv files, list of names of the csv files include the .csv
def concat_frames(Path, file_names):
    count = 0
    frame_dict = {}
    for file in file_names:
        frame_dict[count] = pd.read_csv(Path/file)
        count += 1
    frames = list(frame_dict.values())
    return pd.concat(frames)
```
